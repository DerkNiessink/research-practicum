```python
#df = pd.read_csv("export_0T_2T_angle_0_330deg_2K.dat")
df = pd.read_csv("EXPORT_field _0T_6T_angle_0_330deg_2K.dat")

angles = [30, 45, 90, 150, 210, 225, 270, 330, 0, 15, 60, 75, 105, 120, 135, 165, 180, 195, 240, 255, 285, 300, 315, 345]
positions = {}

for angle in angles:
    positions["{0}deg".format(angle)] = None

position_data = df["Sample Position (degrees)"].values.tolist()

i = 0
j = 0
for position in position_data:
    j +=1
    if j >= len(position_data):
        break
    if abs(position - angles[i+1]) < 1:
        i += 1
    if abs(angles[i] - position) <= 0.1 and positions["{0}deg".format(angles[i])] == None:
        positions["{0}deg".format(angles[i])] = []
    
    positions["{0}deg".format(angles[i])].append((df["Magnetic Field (Oe)"][j], df["Bridge 1 Resistance (Ohms)"][j]))
    

positions.pop('345deg', None)
```

```python
import matplotlib.pyplot as plt


fig = plt.figure()
for angle in positions:
    
    field = []
    resis = []
    for item in positions[angle]:
        field.append(item[0])
        resis.append(item[1])

    plt.plot(field, resis, "o", markersize = 3)
plt.xlabel("Magnetic Field (T)")
plt.ylabel("Resistance (Ohms)")
        
    
```

```python
bc2s =[]
angles_bc2 =[]

for angle in positions:

    # Half of the lists are reversed.
    if positions[angle][0][0] > 10000:
        L = list(reversed(positions[angle]))
    else:
        L = positions[angle]

    maximum = max(positions[angle])
    for item in L:
        # Take the first point with a resistance greater than 0.5* maximum
        if item[1] > 0.5 * maximum[1]:
            bc2s.append(item[0]/ 10000)

            # Convert angle strings into integers.
            if len(angle) == 4:
                angle_int = int(angle[0])
            if len(angle) == 5:
                angle_int = int(angle[0:2])
            if len(angle) == 6:
                angle_int = int(angle[0:3])
            angles_bc2.append(angle_int)
            break

plt.plot(angles_bc2, bc2s, "o")
plt.xlabel("Angle (deg)")
plt.ylabel("Bc2 (T)")
plt.xlim(-15, 350)
plt.ylim(0,0.6)
plt.title("B at 0.5 * max resistance, I = 50 micro-ampere, T = 2K")
```

```python
import numpy as np


def linear(data, percentage):
    # We're searching for this point on a line between two points.
    find = percentage * max(data, key=lambda x: x[1])[1]

    for item in data:

        # Take the point left and right of the initial value.
        if item[1] > find:
            point1 = item
            point2 = data[data.index(point1) - 1]
            break

    # Use the linear function y = ax + b to draw a line from point1 to point2
    a = (point1[1] - point2[1]) / (point1[0] - point2[0])
    b = point1[1] - a * point1[0]

    x_list = []
    y_list = []

    # Make a list of x- and y-coords of the points on the line.
    for x in np.arange(point2[0], point1[0] + 0.01, 0.01):
        x_list.append(x)
        y_list.append(a * x + b)

    # Find the point's y-coord with the smallest difference with the initial value. 
    diff_list = [abs((find - item)) for item in y_list]
    index = diff_list.index(np.min(diff_list))

    return x_list[index], y_list[index]



```

```python
bc2s =[]

try:
    angles.remove(345)
except:
    ValueError

for angle in angles:

    # Half of the lists are reversed
    if positions[f"{angle}deg"][0][0] > 10000:
        L = list(reversed(positions[f"{angle}deg"]))
    else:
        L = positions[f"{angle}deg"]

    coord = linear(L, 0.5)
    bc2s.append(coord[0] / 10000)

plt.plot(angles, bc2s, "o")
plt.xlabel("Angle (deg)")
plt.ylabel("Bc2 (T)")
plt.xlim(-15, 350)
plt.ylim(0, 0.7)
plt.title("B at 0.5 * max resistance, I = 50 micro-ampere, T = 2K")

```
